# text annotation

```python
#draw plot as fig
fig, ax=plt.subplots()
x= np.linspace(0,20,1000)
y=np.cos(x)
ax.plot(x,y)
ax.axis('equal')
ax.annotate('high',xy=(2*np.pi,1),xytext=(10,4),arrowprops=dict(facecolor='red'))
ax.annotate('low',xy=(5*np.pi,-1),xytext=(10,-2),arrowprops=dict(facecolor='red'))
plt.show()
```

# 3D plot

```python
#import mplpt3d tootkit, nump, matplotlib
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
#3d projection 
ax = plt.axes(projection='3d')
z = np.linspace(0,1,100)
x= z*np.sin(25*z)
y= z*np.cos (25*z)
ax.plot3D(x,y,z, 'green')
ax.set_title('het')
plt.show()
```

# pie chart

```python
#creating data set 
country = ['India','Pakistan','Shrilanka','USA','Russia','China']

population =[131,140,130,95,120,125]
explode=(0,0.1,0,0.5,0,0)
plt.pie(population,labels=country,explode=explode,shadow = True)
fig=plt.figure(figsize=(15,12))
plt.show()
```
